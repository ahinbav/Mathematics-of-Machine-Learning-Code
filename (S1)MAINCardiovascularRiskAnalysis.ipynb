```python
import pandas as pd
data=pd.read_csv('FraminghamCardiovascularDataset.csv')
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(20,20))
ax=fig.gca()
data.hist(ax=ax)
plt.show()
```

```python
data.head()
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
X.shape
```

```python
y.shape
```

```python
import numpy as np
np.isnan(X).sum()
```

```python
np.isnan(y).sum()
```

```python
!pip install scikit-learn
```

```python
from sklearn.impute import SimpleImputer

datafiller=SimpleImputer(missing_values=np.nan,strategy='mean')
X=datafiller.fit_transform(X)
```

```python
np.isnan(X).sum()
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
X_train
```

```python
!pip install keras tensorflow
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import SGD
```

```python
model=Sequential()
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
X_train[0]
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=48,epochs=50)
```

```python
history.history
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend({"Loss":1,"Validation Loss":2})
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title('Training and Validation Losses')
```

```python
history.history.keys()
```

```python
loss_legend={'Training_loss':123,'Validation_loss':123}
accuracy_legend={'Training_Accuracy':123,'Validation_Accuracy':123}
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(accuracy_legend)
plt.title('Training and Validation Accuracy')
plt.xlabel("Number of Epochs")
plt.ylabel("Accuracy %")
```
